# Exploratory Data Analysis - AI Art Dataset

```python
from utils import plot_images, load_images, image_stats, pixel_intensity
```

## Loading images

```python
fake_train_dir = './datasets/AI art/train/fake'
real_train_dir = './datasets/AI art/train/real'
fake_test_dir = './datasets/AI art/test/fake'
real_test_dir = './datasets/AI art/test/real'
```

```python
fake_train_images = load_images(fake_train_dir, 5)
real_train_images = load_images(real_train_dir, 5)
fake_test_images = load_images(fake_test_dir, 5)
real_test_images = load_images(real_test_dir, 5)
```

## Visualizing images

```python
plot_images(fake_test_images, 'Fake test images')
plot_images(real_test_images, 'Real test images')
plot_images(fake_train_images, 'Fake train images')
plot_images(real_train_images, 'Real train images')
```

## Statistical summary of images

```python
image_stats(fake_test_images)
image_stats(real_test_images)
image_stats(fake_train_images)
image_stats(real_train_images)
```

## Pixel intensity analysis

```python
pixel_intensity(fake_test_images, title="Fake test images")
pixel_intensity(real_test_images, title="Real test images")
pixel_intensity(fake_train_images, title="Fake train images")
pixel_intensity(real_train_images, title="Real train images")
```
